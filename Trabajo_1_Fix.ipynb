```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargamos el csv
data_frame = pd.read_csv("weatherAUS.csv")
data_frame.head()
```

```python
# Valores nulos
for feature in data_frame.columns:
    print('Total de valores nulos de', feature, '=', data_frame[feature].isna().sum())
```

```python
# X: variable que tendrÃ¡ las columnas para el anÃ¡lisis
X = data_frame.iloc[:, 0:-1]
pd.DataFrame(X).head(10)
```

```python
# Y: variable que tendrÃ¡ la columna a predecir
y = data_frame.iloc[:, -1]
pd.DataFrame(y).head(10)
```

```python
# Dejamos en una variable todas las columnas de tipo object
listObjType = X.select_dtypes(include = "object").columns
print (listObjType)
```

```python
# Transformamos todos los valores en nÃºmeros
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for item in listObjType:
    X[item] = le.fit_transform(X[item].astype(str))

print (X.info())
```

```python
X['Rainfall'] = X['Rainfall'].fillna((X['Rainfall'].mean()))
pd.DataFrame(X).head(100)

#Al ser variable con decimal es necesario utilizar el promedio
```

```python
# Comprobamos que no hay valores nulos
for feature in X.columns:
    print('Total de valores nulos de', feature, '=', X[feature].isna().sum())
```

```python
def getGraficoBarras(label_x, label_y, title, feature):
    x_values = X[feature].unique()
    y_values = X[feature].value_counts().tolist()
    plt.bar(x_values, y_values)
    plt.title(title)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.show()

getGraficoBarras('Indicador de Lluvia Hoy', 'Cantidad de registros', "Valor 0: No, 1: Si", 'RainToday')
```

```python
tipos = X['RainToday'].unique()
total = X['RainToday'].value_counts().tolist()
plt.pie(total, labels=tipos, autopct='%1.1f%%', shadow=True, startangle=90) 
plt.title('DistribuciÃ³n de Lluvia Hoy')
plt.show()
```
